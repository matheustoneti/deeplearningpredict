```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
```

```python
dados = pd.read_csv('data.csv')
```

```python
dados.info()
```

```python
le = preprocessing.LabelEncoder()
```

```python
dados['job'] = le.fit_transform(dados['job'])
dados['marital'] = le.fit_transform(dados['marital'])
dados['education'] = le.fit_transform(dados['education'])
dados['default'] = le.fit_transform(dados['default'])
dados['housing'] = le.fit_transform(dados['housing'])
dados['month'] = le.fit_transform(dados['month'])
dados['poutcome'] = le.fit_transform(dados['poutcome'])
dados['deposit'] = le.fit_transform(dados['deposit'])


```

```python
dados['contact'] = le.fit_transform(dados['contact'])
```

```python
dados['loan'] = le.fit_transform(dados['loan'])
```

```python
dados.head(10)
```

```python
y = dados['deposit']
```

```python
x = dados.drop(['deposit'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size =(0.3))
```

```python
modelo = GradientBoostingClassifier()
```

```python
modelo = modelo.fit(x_train, y_train)
```

```python
accuracy = modelo.score (x_test, y_test)
```

```python
print(accuracy)
```

## Aplicando Deep Learning e comparando os resultados

```python
from sklearn.metrics import accuracy_score
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(units=32, activation='relu', input_dim=len(x_train.columns)))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, epochs=500, batch_size=32)
```

```python
y_hat = model.predict(x_test)
```

```python
resultado = []
for valor in y_hat:
  if valor < 0.5:
    var = 0
  else:
    var = 1
  resultado.append(var)
```

```python
print(resultado)
```

```python
resultado = np.array(resultado)
```

```python
accuracy_score = accuracy_score (y_test, resultado)
```

```python
print('Accuracy com Machine Learning:', accuracy)
print('Accuracy com Deep Learning:', accuracy_score)
```
